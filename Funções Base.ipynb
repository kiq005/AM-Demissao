```python
%matplotlib inline
import math
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

## Tempo de Percurso

```python
x_tp = np.arange(0, 24*60, 0.1)
y_tp = 0.1 + 0.000555556*(x_tp)
plt.plot(x_tp, y_tp)
plt.show()
```

## Salário

```python
def quadratica(x, y_0, y_f, L):
    a = (y_0 - y_f) / L ** 2
    b = -2 * a * L
    return x ** 2 * a + x * b + y_0

x_sal = np.arange(0, 100000) # 100000 seria o limite sup. da BD
y_sal = quadratica(x_sal, .9, .1, 100000)
plt.plot(x_sal, y_sal)
plt.show()
```

## Desempenho

```python
def sigmoid(x, L, k, x_0, y_f):
    return y_f + ((L-y_f)/(1+math.e**(-k*(x-x_0))))
```

```python
def gauss(x, s, m):
    return (1/(s*np.sqrt(2*math.pi)))*math.e**-((x-m)**2/(2*s**2))
```

```python
x_des = np.arange(1, 5, .01)
y_des = sigmoid(x_des, 0.3, -2, 2, 0.1)+.2*gauss(x_des,0.3,2)
#y_des = gauss(x_des, .3, 2)
#y_des = logistica(x_des, .3, -2, 2, .1)
plt.plot(x_des, y_des)
plt.show()
```

## Tempo desde última promoção

```python
x_up = np.arange(0, 20, .1)
y_up = sigmoid(x_up, .9, 2, 5, .1)
plt.plot(x_up, y_up)
plt.show()
```

## Idade


```python
x_id = np.arange(15, 90, .1)
y_id = sigmoid(x_id, 1.5, .6, 23, .1 ) * sigmoid(x_id, .1, .1, 21, 1.5)
plt.plot(x_id, y_id)
plt.show()
```

## Grau de Escolaridade

```python
x_esc = np.arange(0, 9, .1)
y_esc = sigmoid(x_esc, .8, 5.4, 4.3, .3 ) * sigmoid(x_esc, .1, 1.8, 6.6, 1)
plt.plot(x_esc, y_esc)
plt.show()
```

## Tempo Desde Contratação

```python
x_cont = np.arange(0, 10, .1);
y_cont = ( sigmoid(x_cont, .7, 10, .5, .3) + sigmoid(x_cont, 0, 20, .2, .6) ) * sigmoid(x_cont, .1, 1.7, 2.2, 1)
plt.plot(x_cont, y_cont)
plt.show()
```
