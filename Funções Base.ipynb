## Desempenho

```python
def logistica(x, L, k, x_0, y_f):
    return y_f + (L/(1+math.exp(-k*(x-x_0))))
```

```python
def gauss(x, s, m):
    return (1/(s*math.sqrt(2*math.PI)))**(-((x-m)**2)/2*s*s)
```

```python
x_des = np.arange(1, 5, .1)
y_des = logistica(x_des, 0.3, -2, 2, 0.1)+.3*gauss(x_des,0.3,2)
plt.plot(x_des, y_des)
plt.show()
```
