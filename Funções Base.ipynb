## Salário

```python
def func_quadratica(x, yi, yf, l):
    a = (yi - yf) / l ** 2
    b = -2 * a * l
    return x ** 2 * a + x * b + yi

x_sal = np.arange(0, 100000) # 100000 seria o limite sup. da BD
y_sal = func_quadratica(x_sal, .9, .1, 100000)
plt.plot(x_sal, y_sal)
plt.show()
```

## Desempenho

```python
def logistica(x, L, k, x_0, y_f):
    return y_f + (L/(1+math.e**(-k*(x-x_0))))
```

```python
def gauss(x, s, m):
    return (1/(s*np.sqrt(2*math.pi)))*math.e**-((x-m)**2/(2*s**2))
```
